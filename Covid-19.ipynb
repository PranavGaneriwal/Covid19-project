```python
import pandas as pd
import requests
```

## Collecting Data from Wikipedia

```python
url = 'https://en.wikipedia.org/wiki/COVID-19_pandemic_by_country_and_territory'
req = requests.get(url)
```

```python
data_list = pd.read_html(req.text)
```

```python
target_df = data_list[3]
target_df.head()
```

## Data Cleaning

```python
target_df.columns= [ 'Country Name', 'Total Cases', 'Total Deaths', 'Fatality Rate in %', 'Deaths/100,000']
target_df
```

```python
target_df['Country Name'] = target_df['Country Name'].str.replace('\[.*\]','')
target_df['Fatality Rate in %'] = target_df['Fatality Rate in %'].str.replace('%','')
```

```python
target_df['Total Cases'] = pd.to_numeric(target_df['Total Cases'])
target_df['Total Deaths'] = pd.to_numeric(target_df['Total Deaths'])
target_df['Fatality Rate in %'] = pd.to_numeric(target_df['Fatality Rate in %'])
target_df['Deaths/100,000'] = pd.to_numeric(target_df['Deaths/100,000'])
target_df
```

## Exporting the Data

```python
target_df.to_excel(r'Covid19.xlsx')
```
